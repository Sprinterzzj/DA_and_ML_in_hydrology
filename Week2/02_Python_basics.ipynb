## II. Булевы операторы и операторы сравнения

````python
# сравнение булевых переменных
A = True
B = False

print(A and B, A & B)

print(A or B, A | B)

print(not A, not B)
````

````python
# сравнение всего остального и однотипного
print(1 > -1)

print(1 < -1)

print(18 >= 21)

print(21 <= 18) 

print('very long string' > 'short string')

print(99 != 999)
````

````python
n1 = [0, 1]
n2 = [0, 1]
m1 = m2 = [0, 1, 2, 3, 4, 5]

print('равенство n1 и n2:', n1 == n2)
print('идентичность n1 и n2:', n1 is n2)
print()
print('равенство m1 и m2:', m1 == m2)
print('идентичность m1 и m2:', m1 is m2)
````

## IV. Функции

````python
# Стандартные функции

# функция, которая считает суммы квадратов
# четных чисел ряда
def crazy_sum(my_list):
    return sum([i**2 for i in my_list if i%2 == 0])

crazy_sum([i for i in range(1, 5)])
````

````python
# lambda (однострочная функция)
[(lambda x: x**2 - 1)(x) for x in range(1,5)]
````

## Считаем и визуализируем данные архива погоды

Источник данных - http://cliware.meteo.ru/inter/data.html

Пункт наблюдений - Москва-ВДНХ (27612)

После выполнения запроса формируется файл формата .html (переименова в ```moscow.html```)

````python
# импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# парсинг таблицы html
tables = pd.read_html('moscow.html', header=0, index_col="DATE_OBS", parse_dates=True)

# выбор таблицы и параметров
df = tables[0].loc[:, ["TMPMN", "PRECIP"]]

# визуализация данных 2006 года
df['2006'].plot(subplots=True);
````

````python
df2 = tables[0].loc[:, ["TMPMAX", "TMPMIN", "TMPMN"]]

x = df2['2006'].index
y = df2['2006'].TMPMN
y1 = df2['2006'].TMPMAX
y2 = df2['2006'].TMPMIN

fig, ax = plt.subplots()
ax.plot(x, y, color='black')
ax.fill_between(x, y1, y2, facecolor='blue', alpha=0.5)
ax.grid(True)
fig.autofmt_xdate()
````
